# Lecture 31: Learning and Instruction

### **What is Learning?**

Learning is:

* A **cognitive process** and **adaptive skill**
* Leads to **relatively permanent positive changes** in:

  * Cognition and thought processes
  * Behavior
  * Emotional intelligence
  * Social adjustment

### **Layman Perspectives on Learning**

* Quantitative increase in knowledge
* Memorizing and retrieving information
* Learning facts, skills, and problem-solving techniques
* Making sense of abstract concepts (higher cognition)
* Comprehending the world and reinterpreting knowledge

---

## **Learning as a Product**

* Learning outcomes that can be **observed, measured, and assessed**
* Robert Gagné: Learning brings changes in **human disposition or capacity** that:

  * Persist over time
  * Can be evaluated through tools
* Mapped via:

  * Behavioral changes
  * Performance outcomes

---

## **Learning as a Process**

* Continuous internal cognitive and affective processing
* Involves:

  * Understanding
  * Interpretation
  * Application
  * Reflection
  * Problem solving

---

## **Bloom’s Taxonomy (Educational Objectives)**

### **1. Cognitive Domain**

* Mental skills, knowledge processing
* Levels (Revised):

  1. Remember
  2. Understand
  3. Apply
  4. Analyze
  5. Evaluate
  6. Create (highest)

### **2. Affective Domain**

* Emotions, values, attitudes
* Levels:

  * Receiving
  * Responding
  * Valuing
  * Organizing
  * Characterizing (internalizing values)

### **3. Psychomotor Domain**

* Physical and manual skills
* Levels:

  * Reflex movements
  * Fundamental movements
  * Perceptual abilities
  * Physical activities
  * Skilled movements
  * Non-discursive communication (skill + cognition + innovation)

### **Instructional Objective Focus Depends On Subject**

* Engineering: Cognitive + Psychomotor high
* Arts/Literature: Cognitive + Affective more emphasized

---

## **Kolb’s Experiential Learning Cycle**

1. **Concrete Experience**
2. **Reflective Observation**
3. **Abstract Conceptualization**
4. **Active Experimentation**

A continuous loop for real-world learning.

---

## **Dewey’s Reflection Learning Model**

* Reflection begins from **experience**
* Involves:

  * Recalling and detailing events
  * Connecting feelings
  * Using helpful emotions and eliminating obstructive ones
  * Evaluating experiences using goals and prior knowledge
  * Integrating new understanding into cognitive framework

Promotes **self-awareness** and **meta-learning**.

## **Learning and Instruction**

* Instruction = formal term for teaching

# **Major Schools of Learning Orientation**

---

### ✅ Behavioristic Orientation

Focus:

* **Observable behavior**, not internal thoughts
* Learning = change in measurable behavior

Key principles:

* **Stimulus-Response (S-R) bonds**
* Behavior shaped through **environment**
* **Reinforcement** strengthens learning
* **Contiguity**: learning occurs when events are closely paired in time

Emphasis on:

* Reinforcement
* Classroom environment shaping behavior

---

### ✅ Cognitive Orientation

Learning is driven by:

* **Structured, organized instruction**
* Logical sequencing simple to complex
* Highlighting key ideas

Other important principles:

* Perception and attention to stimuli
* Prior knowledge influences new learning
* Acknowledges **individual differences**
* Quick **feedback** promotes improvement

Reinforcement through **knowledge of results** (feedback).

---

### ✅ Humanistic Orientation

Propagated by Carl Rogers

Key aspects:

* Whole-person involvement (cognition + feelings)
* **Self-initiated learning**
* All-pervasive changes in:

  * Behavior
  * Attitude
  * Personality

Learner-centered features:

* Evaluation by the learner
* Meaningful and purposeful learning
* Driven by learner’s interest and needs

Locus of evaluation: **within the learner**
