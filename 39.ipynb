# Lecture – 39: Learning and Pedagogy (Contd.)

## 1. Blended Learning

- **Definition**: Combines online educational materials with face-to-face classroom methods.
- **Key features**:
  - Student control over **time, place, path, and content**.
  - Flexible, convenient, and supports **social learning**.
  - Combines the **best of online and physical classrooms**.
  
### 1.1 Traditional Learning vs. Flip Learning
- **Traditional Learning**:
  - Classroom: lower-order thinking (remembering, understanding, applying)
  - Homework: higher-order thinking (analyzing, evaluating, creating)
- **Flip Learning** (Reverse Learning):
  - Students receive **learning material in advance** (videos, online content)
  - Classroom: higher-order thinking through **discussion, exercises, critical inquiry**
  
### 1.2 Blended Flip Approach
- Combines **blended learning** and **flip learning**
- Key advantages:
  - Flexible environment: students choose **when and where** to study
  - Increases **cognitive flexibility** and **creative thinking**
  - Classroom focuses on **higher-order thinking**, discussion, collaboration, quizzes, group work
  - Online content is **limited, intentional, and controlled** (20–25 min videos)
  - Requires **professionally trained educators**
- **Research Support**:
  - Bergmann & Sams combined blended and flipped approaches
  - Based on **cognitive flexibility theory**
  - Positive impact on teaching-learning outcomes
  - Community of enquiry framework: combines **cognitive presence, social presence, teaching presence**

## 2. Transformative Learning

- **Definition**: Learning that leads to a **complete change in frame of reference**, thought process, mindset, and habits of mind.
- **Focus**:
  - Not just skill transfer or learning outcomes
  - Total transformation of **perspectives, habits, values, ethics**
  
### 2.1 Frame of Reference
- Includes:
  - Habits of mind
  - Personal point of view
- Developed through:
  - Past learning experiences
  - Cultural norms
  - Socialization
  - Personal beliefs and attitudes

### 2.2 Mezirow’s Four Processes of Transformative Learning
1. **Elaborate existing point of view**
2. **Establish new points of view**
3. **Transform previous point of view**
4. **Adopt new habits of mind**

- Learning cycle: **Information → Practice → Reflect → New Information → Practice → Reflection**
- Emphasis on **active and reflective interpretation**:
  - Active: critical evaluation, problem solving, creative thinking
  - Reflective: long-term integration of learning into mindset

### 2.3 Mezirow’s 10 Steps for Transformative Learning
1. **Disoriented dilemma** – encounter conflict or confusion
2. **Self-examination** – analyze internal and external factors
3. **Critical assessment of assumptions**
4. **Recognition of problems** – identify issues and challenges
5. **Exploration of factors** – cause-effect analysis
6. **Planning a course of action** – strategic planning, hypothesis
7. **Provisional trying of new roles/strategies** – trial/pilot phase
8. **Acquisition of knowledge and skills** – realization of action plan
9. **Building confidence and competence** – evaluate effectiveness
10. **Reintegration** – adopt new perspective into habits and attitude

- Outcome: **students adopt new perspectives, become innovative and reflective**, and integrate transformative learning into personal and professional life.
