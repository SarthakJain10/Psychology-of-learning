# Lecture – 35: Learning and Instruction (Contd.)

## Robert Gagne's Theory of Learning and Instruction

- **Overview**:  
  - Gagne is primarily a **behaviorist**, but his instructional model blends **behaviorism** and **cognitive concepts**.  
  - Learning is seen as a result of **observable behavioral changes** and **cognitive effort**.  
  - Learning outcomes are **measurable** and linked to **educational goals**.

- **Key Focus**:  
  - Curriculum design should **avoid overloading memory**.  
  - Emphasis on **thinking and remembering processes**.  

### Three Elements of Gagne’s Theory
1. **Taxonomy of Learning Outcomes**  
2. **Conditions Necessary to Achieve Learning Outcomes**  
3. **Nine Events of Instruction** – guides the instructional process.

### Taxonomy of Learning (5 Categories)
- Each category requires **different types of instruction**:
  - Verbal information
  - Intellectual skills
  - Cognitive strategies
  - Attitudes
  - Motor skills

- **Domains Covered**:
  - **Cognitive**: Intellectual skills, memory, understanding  
  - **Affective**: Attitudes, motivation  
  - **Psychomotor**: Motor skills

### Mandatory Teacher Tasks
1. Assure **prerequisite skills**  
2. Supply **internal and external instructional conditions**  
3. Determine the **type of learning**  

### Nine Events of Instruction
1. Gaining attention  
2. Informing learners of objectives  
3. Stimulating recall of prior learning  
4. Presenting the stimulus  
5. Providing learning guidance  
6. Eliciting performance  
7. Providing feedback  
8. Assessing performance  
9. Enhancing retention and transfer  

> **Summary**: Gagne’s model integrates **behavioral outcomes** and **cognitive strategies** for effective instruction.

## Jerome Bruner's Guided Discovery Learning

- **Overview**:  
  - Learner is **active** and existing **schema** is important.  
  - Emphasizes **discovery with guidance**.  
  - Goal: Develop **autonomous learners**.

- **Intellectual Development Stages**:
  1. **Enactive Stage**: Action-based learning  
  2. **Iconic Stage**: Image-based learning  
  3. **Symbolic Stage**: Language-based learning  

- **Culturally Invented Technologies**:  
  - Include **language, abstract notions**, and tools (not just computers/TVs).  
  - Mediate interaction with environment and enhance cognitive growth.

### Modes of Representation
- **Enactive**: Learning through actions and experiences  
- **Iconic**: Learning through images, visualizations  
- **Symbolic**: Learning through language, symbols  

**Progression**: Enactive → Iconic → Symbolic for effective instruction.

### Key Implications
- Students should **discover relationships** between concepts themselves.  
- Teachers provide **cognitive scaffolding**: gradual guidance and support.  
- The **level of instruction** must match the learner's **developmental stage**.

## Piaget’s Cognitive Development and Learning

- **Overview**:  
  - Focused on **intellectual/cognitive development** from birth to adolescence.  
  - Children learn through **exploration, assimilation, and accommodation**.

- **Learning Cycle**:
  1. **Exploration**  
  2. **Concept Introduction**  
  3. **Concept Application**  

- **Mechanism**:
  - **Assimilation**: Integrating new info into existing schemas  
  - **Accommodation**: Modifying schemas based on new info  
  - **Equilibrium**: Balanced cognitive state  
  - **Disequilibrium**: Encountering conflict → prompts learning  

### Developmental Stages
1. **Sensory-Motor** (0–2 yrs): Learn through physical actions  
2. **Pre-Operational** (2–7 yrs): Symbolic thinking begins  
3. **Concrete Operational** (7–11 yrs): Logical thinking on concrete objects  
4. **Formal Operational** (11+ yrs): Abstract reasoning and hypothesis formation  

- **Key Principle**: Learning is **active and self-constructed**, guided by personal experience.  
- **Teacher’s Role**: Facilitate exploration without over-intervention.

### Summary of Instructional Approaches
| Theorist | Key Idea | Instructional Implication |
|-----------|----------|---------------------------|
| **Gagne** | Behavioral + cognitive outcomes | Use 9 events of instruction, ensure prerequisites, measurable outcomes |
| **Bruner** | Guided discovery, active learner | Use enactive → iconic → symbolic progression, scaffold learning |
| **Piaget** | Cognitive development stages | Align instruction with developmental stage, allow exploration, assimilation & accommodation |
