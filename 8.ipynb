# Lecture 8: Major Theories of Learning (Contd.)

## 🌐 Cognitive Revolution: Beyond Gestalt

* Gestalt psychology marked the shift from behaviorism to **cognitive approaches**.
* Two directions emerged:

  * **Neuropsychology (Brain metaphor):** Studies brain structure, lobes, neural centers.
  * **Information Processing (Computer metaphor):** Human mind as a processor of input → processing → output.

---

## 🧠 Information Processing Approach

Human cognition compared to a computer system:

| Stage                       | Description                                          |
| --------------------------- | ---------------------------------------------------- |
| **Sensory Memory**          | Brief store of incoming sensory information          |
| **Short-Term Memory (STM)** | Active processing; requires rehearsal                |
| **Long-Term Memory (LTM)**  | Knowledge repository; retained via deeper processing |

Key processes during transfer to LTM:

* **Encoding**
* **Elaboration**
* **Practice**
* **Cognitive strategies**

Outcome of processing becomes **behavioral response** guided by planning and decision-making (frontal lobe involvement).

> Emerged as a response to the limitations of behaviorism and suggests mind has **limited capacity** that can be improved through strategic learning.

## 🧩 Introduction to Intelligence

* Intelligence enables individuals to:

  * Understand complex ideas
  * Adapt effectively
  * Learn from experience
  * Think rationally and act purposefully

### Measurement Tradition

| Psychologist    | Contribution                                    |
| --------------- | ----------------------------------------------- |
| **Binet-Simon** | First IQ scale; mental age vs chronological age |
| **Wechsler**    | Intelligence scales for children and adults     |

**IQ Formula:**
`IQ = (Mental Age / Chronological Age) × 100`

Modern view expands beyond IQ to include **Emotional Intelligence (EQ)**.

> “IQ may get you hired, EQ helps you rise.”

---

## 🌈 Gardner’s Theory of Multiple Intelligences

Against the idea of a single general intelligence (**g-factor**).

Initial **7 intelligences**:

1. **Linguistic**
2. **Logical-Mathematical**
3. **Spatial**
4. **Bodily-Kinesthetic**
5. **Musical**
6. **Interpersonal**
7. **Intrapersonal**

Later additions:
8. **Naturalistic** — Understanding nature and ecology
9. **Existential** — Awareness of deeper philosophical and psychological life questions

> Each individual possesses a **unique profile** of intelligences.

---

## 🔺 Sternberg’s Triarchic Theory of Intelligence

Three **interrelated** types making up Successful Intelligence:

| Type                        | Focus                      | Real-Life Role             |
| --------------------------- | -------------------------- | -------------------------- |
| **Analytical Intelligence** | Problem-solving, reasoning | Academic success, IQ tests |
| **Creative Intelligence**   | Novel ideas, innovation    | Creativity, adaptability   |
| **Practical Intelligence**  | Application of knowledge   | Everyday problem-solving   |

---

### Structure: Sub-Theories & Components

| Sub-Theory       | Focus                        | Components                                                                 |
| ---------------- | ---------------------------- | -------------------------------------------------------------------------- |
| **Componential** | Cognitive mechanisms         | Meta-components (planning), Performance (execution), Knowledge-acquisition |
| **Experiential** | Role of experience           | Novelty handling → Automation                                              |
| **Contextual**   | Interaction with environment | Adaptation, Selection, Shaping                                             |

Success depends on:

* Motivation
* Environment
* Use of personal abilities

---

## ✅ Summary Highlights

* Cognitive psychology shifted learning focus to **mental processing**.
* Intelligence is **multi-dimensional**, dynamic, and context-dependent.
* Multiple frameworks help us understand how humans **learn, adapt, and excel**.
