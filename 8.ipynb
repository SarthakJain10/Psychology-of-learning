## ðŸ§  Information Processing Approach

Human cognition compared to a computer system:

| Stage                       | Description                                          |
| --------------------------- | ---------------------------------------------------- |
| **Sensory Memory**          | Brief store of incoming sensory information          |
| **Short-Term Memory (STM)** | Active processing; requires rehearsal                |
| **Long-Term Memory (LTM)**  | Knowledge repository; retained via deeper processing |

Key processes during transfer to LTM:

* **Encoding**
* **Elaboration**
* **Practice**
* **Cognitive strategies**

Outcome of processing becomes **behavioral response** guided by planning and decision-making (frontal lobe involvement).

> Emerged as a response to the limitations of behaviorism and suggests mind has **limited capacity** that can be improved through strategic learning.
