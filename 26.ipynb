## üåü What is Emotional Intelligence (EI)?

Emotional Intelligence refers to the **ability to identify, understand, manage, and express** one‚Äôs own emotions and the emotions of others.

EI was originally embedded within **Social Intelligence**, but became widely recognized through researchers like:

* **Daniel Goleman** (popularized EI in 1995)
* **Peter Salovey & John Mayer** (1990 ‚Äì EQ Theory)

EI also known as:

* **EQ ‚Äì Emotional Quotient**

## üí° Key Components of Emotional Intelligence

| Component           | Meaning                                          | Involves                                                     |
| ------------------- | ------------------------------------------------ | ------------------------------------------------------------ |
| **Self-Awareness**  | Recognizing one's own emotions                   | Emotional literacy, identifying causes & effects of feelings |
| **Self-Regulation** | Controlling & expressing emotions appropriately  | Monitoring reactions, reducing impulsivity                   |
| **Motivation**      | Using emotions to drive toward goals             | Intrinsic motivation, perseverance, strategic planning       |
| **Empathy**         | Understanding & responding to emotions of others | Authentic emotional attunement, helping & comforting         |
| **Social Skills**   | Managing relationships effectively               | Communication, collaboration, conflict resolution            |
