# Major Theories of Learning (Contd.)

## ‚úÖ Recap: Behaviorist Learning Theories

Classical Conditioning and Operant Conditioning influence classroom behavior by focusing on:

* **Participation**
  Requires attention, motivation, and mental readiness.

* **Feedback**
  Knowledge of results improves learning and performance.

* **Reinforcement**
  Keeps learners motivated and engaged in tasks.

* **Stimuli for Behavior Control**
  Effective stimuli help modify behavior and promote positive learning tendencies.

* **Learning Measurement**
  Behavior is observable and measurable, so tools can track classroom learning success.

* **Forgetting & Exam Timing**
  Delayed exams may reduce retention and motivation.

* **Intermittent Practice**
  Distributed learning over time improves long-term retention.

## üß† Cognitive Side of Behaviorism

Even within operant conditioning, cognitive interpretations exist.

### Learned Helplessness

* Proposed by **Martin Seligman**
* Occurs when individuals believe they have no control over outcomes.
* Leads to **inactivity**, **demotivation**, and a **sense of incompetence**.

**Educational Implication:**
Prevent recurring failures, provide support, build self-efficacy.

## üëÄ Observational Learning (Albert Bandura)

Learning happens by **watching others**, even without reinforcement.

### Key Processes

| Process          | Meaning                       |
| ---------------- | ----------------------------- |
| **Attention**    | Focus on the model            |
| **Retention**    | Store observed acts in memory |
| **Reproduction** | Convert memory into actions   |
| **Motivation**   | Desire to imitate             |

### Why imitate a model?

* Extrinsic motivation: model is rewarded or admired
* Intrinsic motivation: learner likes the model or behavior

Example: Young athletes copying successful sports stars.

‚úÖ Observational learning + classical + operant conditioning form the **Behavioristic School**.

## üîÑ Shift Toward Cognitivism

Behaviorism faced criticism for:

* Narrow focus on **observable behaviors**
* Ignoring **thinking, creativity, language**
* Animal studies not always applicable to humans

This sparked emergence of **cognitive psychology**.

---

## üß© Cognitivism: Core Ideas

* Learning involves **internal mental processes**
* Mind works like an **information processor**

  * Receives ‚Üí Organizes ‚Üí Stores ‚Üí Retrieves information

### Cognitive Activities

* Attention
* Memory
* Thinking & reasoning
* Creativity
* Problem-solving

Learners **actively construct** knowledge by linking:

> New learning + Prior knowledge ‚Üí Meaningful learning
