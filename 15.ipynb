# Cognitive Load

- **Definition:** Cognitive load refers to the extra burden on working memory when processing information.
  - Can arise from:
    - Task difficulty
    - Content complexity
    - Problem-solving efforts
- Goal:
  - Reduce cognitive load
  - Enhance working memory processing and retention in long-term memory (LTM)

---

## Working Memory & Schema

- **Schemas:** Concepts or ideas formed from experiences, stimuli, or environment.
  - Help **compress information** into manageable units.
  - Reduce working memory load.
  - Facilitate **long-term memory storage**.

- **Working memory limitations:**
  - Visual Sketchpad
  - Phonological Loop
  - Episodic Buffer
- Effective processing in working memory is crucial for LTM retention.

---

## Cognitive Load Theory

- Assumes:
  - **Limited working memory**  
  - **Unlimited long-term memory**
- **Key points:**
  - Processing must be effective for long-term retention.
  - Cognitive load arises during working memory operations.
- **Factors influencing cognitive load:**
  - **Causal factors**: Subject characteristics, task complexity, environment, and their interactions.
  - **Assessment factors**: Mental load, mental effort, and performance outcomes.

---

## Causal Factors

1. **Subject characteristics**: Cognitive ability, skills, prior knowledge.
2. **Task complexity**: Difficulty level of problems or content.
3. **Environment**: Noise, distractions, and learning conditions.
4. **Mutual interactions**: Combined effect of subject, task, and environment on the learner.

---

## Assessment Factors

- **Mental load**: Imposed by task, subject, and environmental demands.
- **Mental effort**: Cognitive capacity allocated by the learner to the task.
- **Performance**: Reflects mental load and effort on task outcome.
- **Processing types:**
  - **Control processing**: Effortful, conscious processing
  - **Automatic processing**: Effortless, mastered through repetition

---

## Types of Cognitive Load

1. **Intrinsic Load**
   - Inherent complexity of the task or subject.
   - Example: Complexity of a math problem.
2. **Extraneous Load**
   - Caused by instructional design, environmental factors, or distractions.
   - Needs minimization for efficient learning.
3. **Germane Load**
   - Constructive load required for schema construction and effective information processing.
   - Focuses on deep learning and long-term retention.

**Total Cognitive Load** = Intrinsic Load + Extraneous Load + Germane Load

---

## Strategies to Reduce Cognitive Load

- **One concept per page**: Summarize key concepts clearly.
- **Remove irrelevant items**: Focus on essential information.
- **Chunking & categorization**: Group related information for easier recall.
- **Consistent structure**: Develop a learning style or format for summarization.
- **Link to learning objectives**: Connect lessons to goals and outcomes.

---

## Memory Protection & Brain Health

- Avoid stress, alcohol, drugs, neurotoxins.
- Maintain nutrition, sleep, and oxygenation.
- Engage in deep breathing, meditation, and yoga.
- Stimulate brain constructively:
  - Puzzles, problem-solving, constructive tasks
  - Avoid unnecessary information overload
- Unlearn redundant or outdated information
- Maintain self-efficacy and confidence
- Stay mentally active throughout life to prevent cognitive decline

---

## Learning & Memorization Techniques

- Use mnemonic devices and coding systems.
- Self-repetition and rehearsal.
- Chunking and grouping of information.
- Concept mapping or mental mapping (one-page sketches).
- Testing and self-assessment:
  - Peer feedback
  - Online tests
- Attach personal cues or interests to aid memory:
  - Music, hobbies, personal experiences

---

## Summary

- Cognitive load management is essential for effective learning.
- Reduce **extraneous load** and maximize **germane load** for deep processing.
- Use **schemas, chunking, consistent structure, and mnemonics**.
- Protect and stimulate the brain through healthy habits and mental engagement.
- Goal: Enhance working memory efficiency and long-term retention.
