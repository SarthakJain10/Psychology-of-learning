## Working Memory Model

- **Short-term memory** is now conceptualized as **working memory**: a temporary workbench for actively processing information.
- Working memory allows us to work on information until tasks are completed and challenges resolved.

### Components of Working Memory

1. **Central Executive**
   - Acts as the decision-maker.
   - Directs attention, executes plans, and issues instructions to other components.
   - Coordinates multiple cognitive processes simultaneously.
   
2. **Visual Sketchpad**
   - Holds and manipulates **visual information** such as images and spatial data.
   
3. **Phonological Loop**
   - Processes **auditory/verbal information**.
   - Uses rehearsal mechanisms to prevent rapid decay of information.

4. **Episodic Buffer**
   - Integrates visual, auditory, and spatial information.
   - Acts as an attentional system to combine multiple sources of information.
   - Interfaces with **Long-Term Memory (LTM)** to store episodic memories.

---

## Assumptions of Working Memory

1. **Dual Channel Processing**
   - Information is processed via two separate channels: **visual/pictorial** and **auditory/verbal**.
   - Known as **Dual Coding Theory** (Paivio & Baddeley, 1998).

2. **Limited Capacity**
   - Working memory has a finite ability to hold information.
   - Example: Visual memory digit span = ~5 Â± 2 items; auditory memory has similar constraints.

3. **Active Processing**
   - Meaningful learning requires **substantial cognitive effort** to process verbal and visual information.
   - Active processing enables retention in LTM.

---

## Principles of Working Memory Functioning

- Visual and auditory information systems are **relatively independent**.
- **Simultaneous/mixed-mode presentation** (multimedia) enhances learning by reinforcing both channels.
- Working memory predicts performance in complex cognitive tasks such as:
  - Planning
  - Problem-solving
  - Decision-making

---

## Knowledge Types in Working Memory

1. **Procedural Knowledge**
   - "Know-how" information or action-oriented skills.

2. **Semantic Knowledge**
   - Abstract or meaningful interpretation of information.

---

## Function of Each Component

| Component          | Function                                                                 |
|-------------------|-------------------------------------------------------------------------|
| **Central Executive** | Supervises all processes, decides relevance, suppresses irrelevant info, coordinates subsystems |
| **Visual Sketchpad**  | Maintains and manipulates visual and spatial information               |
| **Phonological Loop** | Stores and rehearses auditory/verbal information to prevent decay    |
| **Episodic Buffer**   | Temporarily integrates visual, auditory, and spatial info; interfaces with LTM |

---

## Working Memory as a Workspace

- Acts as a **temporary workspace** for holding and manipulating information.
- Supports complex cognitive tasks:
  - Reasoning
  - Analysis
  - Evaluation
  - Critical and reflective thinking
- Interfaces with **Long-Term Memory** for storage of meaningful information.

---

## Intelligence Systems Related to Working Memory

1. **Fluid Intelligence**
   - Basic biological cognitive abilities.
   - Initial raw processing capacity.

2. **Crystallized Intelligence**
   - Developed with **education, practice, and training**.
   - Strengthened and consolidated knowledge.

---

## Strategies to Enhance Working Memory and Learning

- Mnemonic strategies
- Cognitive maps / Mind mapping / Concept mapping
- Sketching, drawing, and visual summarization
- Repetition and rehearsal
- Application of reflective and critical thinking

---

## Summary

- Working memory is a **limited-capacity, multi-component system**.
- Essential for **meaningful learning, reasoning, comprehension, and analysis**.
- Complex tasks require a mental workspace to hold and manipulate information.
- Continuous interface exists between **working memory** and **long-term memory**.
- Effective strategies can **enhance working memory capacity** and improve long-term retention.
