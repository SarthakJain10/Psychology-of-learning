## üß† Atkinson‚Äôs Achievement Motivation Theory

Focus:
‚Ä¢ Interaction between **individual characteristics** and the **need to achieve**
‚Ä¢ Reasons vary: power, social service, prestige, self-efficacy, money, etc.

Two key tendencies:

1. **Motive to achieve success** (drives persistence and goal pursuit)
2. **Motive to avoid failure** (driven by fear of negative outcomes such as punishment or self-criticism)
