## Problem-Based Learning (PBL) vs Traditional Learning
- Traditional learning emphasizes prescribed learning, memorization, and reproduction.
- **Problem-Based Learning (PBL)** is a pedagogical approach rooted in constructive learning theories.
- **Key characteristics of PBL:**
  - Immersive and experiential learning
  - Engagement with real-life problems
  - Development of multiple skills: cognitive, social, emotional, and learning skills

### Skills Developed Through PBL
1. **Self-Directed Learning**  
   - Learners explore problems and formulate strategies independently.
2. **Critical Thinking Skills**  
   - Evaluation, reflection, and planning before taking action.
3. **Cooperative Learning Skills**  
   - Team discussions, peer tutoring, sharing perspectives.
   - Development of social and emotional skills.
4. **Communication Skills**  
   - Mutual understanding, empathy, and expressing ideas clearly.
5. **Higher-Order Thinking Skills**  
   - Reflective, metacognitive, creative, and innovative thinking.
6. **Information Management Skills**  
   - Ability to gather, utilize, and apply knowledge from multiple sources.
7. **Decision-Making Ability**  
   - Strategic planning and evaluation of effective solutions.

## Concept Mapping
- **Definition:** Visual representation of knowledge; shows relationships between concepts.
- **Components:**
  - Nodes: represent concepts
  - Links: directional connections between nodes
  - Linking phrases: describe semantic relationships
- **Purpose:**
  - Acts as an advance organizer for lessons
  - Enhances meaningful learning
  - Supports memory retention in long-term memory
  - Helps identify incorrect or incomplete ideas
  - Promotes integrative and strategic learning

### Steps to Build a Concept Map
1. Identify the **main idea** or topic.
2. Determine **key concepts**: functions, operations, applications.
3. Link concepts hierarchically or in networks (spider maps, flowcharts, circular, 3D structures).
4. Use colors for different relationships or categories.
5. Revise, expand, and elaborate periodically.

### Theoretical Framework
- Based on **David Ausubel's Assimilation Theory**:
  - Concept maps act as **advance organizers**.
  - Facilitate integration of new concepts into existing cognitive structures.
- Supports **cognitive mapping**:
  - Assimilates new information without altering existing knowledge
  - Enhances long-term memory and overall cognitive performance

### Benefits of Concept Mapping
- Organizes knowledge holistically
- Enhances comprehension, integration, and recall
- Facilitates brainstorming and idea generation
- Acts as both a learning and evaluation tool

## Cognitive Maps
- Cognitive maps are **mental representations of knowledge**.
- Developed through concept mapping:
  - Assimilation of new concepts into existing schemas
  - Strengthens cognitive structure and long-term memory
- Can be visualized using software tools for better understanding and integration.
