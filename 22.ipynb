## 1Ô∏è‚É£ Drive Reduction Theory

**Proposed by:** Clark Hull (1943)
**Core Idea:**
Biological or psychological **needs create drives** ‚Üí Drives push behavior ‚Üí Behavior reduces the drive ‚Üí Reduction strengthens behavior.

| Concept       | Explanation                               |
| ------------- | ----------------------------------------- |
| Homeostasis   | Balanced state, no unmet needs            |
| Drive         | Internal tension (e.g., hunger, thirst)   |
| Activation    | Goal-directed behavior to reduce drive    |
| Reinforcement | Drive reduction strengthens that behavior |

### Types of Drives

| Level     | Drive Type    | Examples                      | Importance                 |
| --------- | ------------- | ----------------------------- | -------------------------- |
| Primary   | Biological    | Food, water, reproduction     | Survival                   |
| Secondary | Psychological | Love, approval, recognition   | Mental well-being          |
| Tertiary  | Social        | Status, belonging, leadership | Social identity & prestige |

‚û°Ô∏è Hierarchical progression: Biological ‚Üí Psychological ‚Üí Social

## 2Ô∏è‚É£ Incentive Theory

**Based on:** Operant Conditioning (Skinner)

**Core Idea:**
People act to gain **rewards** and avoid **punishments**. Motivation is driven by **external incentives**, not internal drives.

### Types of Reinforcement

| Type                   | What happens?               | Effect                        |
| ---------------------- | --------------------------- | ----------------------------- |
| Positive Reinforcement | Add pleasant reward         | Increase desired behavior     |
| Negative Reinforcement | Remove unpleasant condition | Increase desired behavior     |
| Punishment             | Add unpleasant consequence  | Decrease undesirable behavior |

‚úÖ Works especially well for shaping children‚Äôs behavior
‚úÖ Links to **extrinsic motivation**

Example:
Child receives a chocolate after finishing dinner ‚Üí eating behavior strengthened.

## 3Ô∏è‚É£ Arousal Theory

**Proposed by:** Henry Murray (1938)
Later expanded (Yerkes‚ÄìDodson Law)

**Core Idea:**
People seek an **optimal level of arousal**. Too little: boredom & low performance. Too much: anxiety & poor performance.

### Key Human Needs (Murray)

* **Need for Achievement**
* **Need for Power**
* **Need for Affiliation**

These motivate action to maintain comfortable arousal levels.

---

### üìà Yerkes‚ÄìDodson Law (Arousal vs Performance)

Performance depends on arousal level:

| Arousal Level | Performance Outcome              |
| ------------- | -------------------------------- |
| Very Low      | Poor (boredom, low engagement)   |
| Optimal       | Best performance ‚úÖ               |
| Too High      | Poor again (stress, nervousness) |

Terminology:

* **Eustress**: Positive, motivating stress
* **Distress**: Too much stress ‚Üí performance drops

---

### ‚ú® Characteristics of Arousal

* Subjective and differs by personality
* Can vary with context, importance, and mood
* Requires enough activation to engage in learning

Examples:

* Feeling sleepy while studying ‚Üí drink coffee to raise arousal
* Some enjoy nightlife (high-arousal preference) while others avoid crowds
