## ‚úÖ Types of Learning

### 1Ô∏è‚É£ Non-Associative Learning

Learning from a single repeated stimulus without forming associations.

#### a) **Habituation**

* Response to a repeated stimulus gradually decreases
* Becomes part of regular experience
* **Example:** Ignoring the ticking sound of a clock after some time

#### b) **Sensitization**

* Response increases when exposed to a strong or purposeful stimulus
* Often induced through medication or external agents
* **Example:** Increased alertness after a loud noise or drugs affecting behavior

---

### 2Ô∏è‚É£ Associative Learning

* Links two or more events together
* Forms meaningful connections
* Occurs through coincidence, exposure, and experience
* **Example:** Associating lightning with thunder

---

### 3Ô∏è‚É£ Meaningful Learning

Connecting new information with **prior knowledge** to create deeper understanding.

* Promoted by **David Ausubel**
* Enhances relevance, cognition, and self-construction of knowledge
* Opposite of rote memorization
* Helps in conceptual understanding

> ‚ÄúSignificant learning occurs when new information is linked to existing cognitive structure.‚Äù

---

### 4Ô∏è‚É£ Cooperative (Collaborative) Learning

Learning through **group interaction**, communication, idea sharing, and peer support.

Promoted by **John Dewey**

#### Key Elements:

* **Heterogeneous groups** (diversity increases insight)
* **Positive interdependence** (mutual benefit & support)
* **Effective communication** and feedback
* **Individual responsibility** (everyone contributes)

Benefits:

* Clarifies doubts, encourages creativity
* Builds social and academic skills

---

### 5Ô∏è‚É£ Emotional Learning

Learning influenced by emotional experiences.

* Strong affective stimuli impact behavior
* Used in therapy to modify emotions and behavior
* **Systematic desensitization** reduces irrational fears and phobias
* Common in **clinical and counseling** contexts

---

### 6Ô∏è‚É£ Observational Learning

Learning by watching and imitating others.
Part of **Albert Bandura‚Äôs Social Learning Theory**.

Process:

* Observe ‚Üí Evaluate ‚Üí Imitate ‚Üí Reinforce
* Role models shape behaviors, manners, and habits
* Helps develop social skills and discipline

---

### 7Ô∏è‚É£ Learning by Discovery

Learning through exploration rather than prescribed instruction.

Promoted by **Jerome Bruner**

Features:

* Learner investigates concepts independently
* Teacher acts as a **facilitator**
* Encourages curiosity, creativity, problem solving
* Strengthens **metacognition**: knowing how we think and learn

Used in:

* Problem-based learning
* Laboratory or practical tasks

---

### 8Ô∏è‚É£ Rote Learning

* Memorization by repetition
* Useful for foundational knowledge like formulas or tables
* Not effective for complex or conceptual learning
* Plays limited role in higher-order cognitive development
