## Social and Situational Orientation to Learning

* Focus: Learning occurs within **social and situational contexts**.
* Key aspects:

  * Observational learning (modeling behavior)
  * Group learning and interaction
  * Situated learning: Knowledge acquisition depends on **social relationships and co-participation**.

### Observational Learning

* Initiated by **Alfred Bandura**.
* Learning occurs by **observing others** in a social context.
* Emphasizes **group discussions, interactions, and collaboration**.

### Situated Learning

* Learning is **contextual** and occurs in communities of practice.
* Focuses on **relationships, participation, and collaboration** rather than just content.
* Examples:

  * Family learning
  * Classroom learning
  * Workplace learning

## Constructive and Social-Constructive Orientation

### Constructive Learning

* Learner is **active in creating knowledge**.
* Knowledge is built through:

  * Individual experiences
  * Interpretation of interactions
* Mind is an **active participant** in learning.
* Learning involves **cognitive reframing**, creating new schemas and beliefs.

### Social-Constructive Learning

* Knowledge is constructed **socially**, through collaborative learning.
* Focus on:

  * Peer learning
  * Cooperative learning
  * Reciprocal exchange of ideas
* Social context plays a key role in shaping understanding.

## Learning Process

### Components

1. **Inputs:** Data, information from environment and social context
2. **Processes:** Mental mechanisms like acquisition, retention, recall, application
3. **Outcomes:** Competencies, knowledge, skills, attitudinal changes

### Factors Affecting Learning

1. **Individual Variables**

   * Motivation, interest, perception
   * Past experiences, intelligence, endurance, patience
   * Personal preferences, ego, and personality
2. **Task-Related Variables**

   * Objectives, content, instructional goals
   * Trainer competence and experience
   * Communication, rationality, task load, interventions
3. **Environmental Variables**

   * Learning context: classroom, field-based, infrastructure
   * Ecology, ambience, noise, lighting, temperature
   * Ergonomics, facilities, hands-on exposure, service quality

---

## Types of Learning

### Active vs Passive

* **Active Learning:** Hands-on, experiential, learning by doing, experimenting
* **Passive Learning:** Teacher-centered, listening, memorizing, observing

### Shallow vs Deep

* **Shallow:** Superficial, surface-level understanding
* **Deep:** Thorough understanding, problem-solving, self-evaluation

---

## Learning Styles

1. **Visual Learners**

   * Prefer **diagrams, charts, graphs, concept maps**
   * Learn better through **visual representation** of ideas

2. **Auditory Learners**

   * Prefer **listening, lectures, group discussions**
   * Benefit from **repeating lessons, audio materials**

3. **Kinesthetic Learners**

   * Learn best through **hands-on experience and doing**

4. **Reading & Writing Learners**

   * Prefer **text-based learning**, reading and summarizing in own words
   * Excel in **written assignments and quizzes**

5. **Logical/Analytical Learners**

   * Focus on **cause-effect, patterns, interrelationships**
   * Motivated by **problem-solving and analytical tasks**

6. **Social/Linguistic Learners**

   * Learn through **peer interaction, discussions, storytelling**
   * Prefer **group work, collaborative learning**

7. **Solitary Learners**

   * Prefer **individual learning and self-study**
   * Perform better through **journaling, diary writing, independent tasks**

8. **Nature Learners**

   * Learn best in **natural environments**
   * Prefer **hands-on, outdoor learning experiences**
   * Example: Gurukul teaching system (learning under trees)
