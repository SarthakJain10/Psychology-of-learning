## PASS Theory of Intelligence

**Developed by**: Das (India), Naglieri, and Kirby (Canada) – 1994
**PASS** stands for:

* **P**: Planning
* **A**: Attention (Arousal)
* **S**: Simultaneous Processing
* **S**: Successive Processing

It represents a cognitive approach focusing on **how information is processed** rather than just the outcome (like IQ score).

---

### Functional Units of PASS Theory

| Functional Unit                            | Key Role                                              | Brain Area                        |
| ------------------------------------------ | ----------------------------------------------------- | --------------------------------- |
| **Attention & Arousal**                    | Directs focus to relevant stimuli                     | Brain stem & related areas        |
| **Simultaneous and Successive Processing** | Coding, organizing & transforming information         | Parietal-Occipital-Temporal lobes |
| **Planning**                               | Decision-making, strategy, execution, self-monitoring | Frontal & prefrontal cortex       |

---

### Processing Types

| Processing Type  | Usage                                  | Example                                  |
| ---------------- | -------------------------------------- | ---------------------------------------- |
| **Simultaneous** | Combining multiple inputs into a whole | Listening, observing and acting together |
| **Successive**   | Sequential information handling        | Memorizing mobile numbers or dates       |

---

### Key Points

* Inspired by **Alexander Luria’s Neuropsychological Theory**
* Focuses on **psychological processes** as building blocks of intelligence
* Operates on a **knowledge base** foundation: cultural, social, academic information

---

## Planning Component

Includes:

* Strategy generation
* Evaluation & execution
* Self-monitoring & impulse control
* Use of prior knowledge & skills
* Higher-order thinking (analysis, synthesis)

Distinguishes humans from other animals due to complex frontal lobe functioning.

---

## Attention Component

* Related to orientation and cortical arousal
* Must reach optimum level for further processing
* Regulated by intentions, goals, emotions, mood, and skills
* Longer attention required for tasks with high difficulty/resilience

---

## Cognitive Assessment System (CAS)

A **standardized test battery** developed from PASS:

* Measures attention, simultaneous, successive, and planning processes
* Provides diagnostic insights
* Supports **remedial interventions**
* Available in multiple languages including **Indian languages**
* Useful for:

  * Learning disabilities
  * Slow learners
  * Reading and mathematical difficulties

# Brain-Based Learning

A **neuropsychological perspective** on learning
Explores:

* Brain structure and function
* Neural networks and electrochemical changes
* Brain hemispheric functions

**Key idea**: Learning builds and strengthens neural connections.

### Emergence & Contributors

* Eric Jensen (2000)
* Kane & Kane (2002) further expanded the theory

---

### Factors Ensuring Meaningful & Permanent Learning

1. **Relaxed Alertness**
   Calm yet attentive state boosts learning

2. **Orchestrated Immersion**
   Integrated and interconnected learning materials allow deeper engagement

3. **Active Processing**
   Application, reflection, mental engagement strengthen neural pathways
   “An idle brain is the devil’s workshop” highlights need for activation

---

### Principles of Brain-Based Learning

* Brain is a **parallel processor**
* Learning engages **entire physiology**
* Search for meaning occurs through **patterning**
* Emotions strongly influence pattern formation
* Brain processes both **whole and parts simultaneously**
* Environmental factors affect learning
* Conscious + unconscious processes contribute
* Learning involves **STM and LTM**
* Prior knowledge supports new learning
* Brain stores data in multiple places
* Regular reflection improves retention and retrieval

---

### Factors Influencing Brain-Based Learning

* Organic and biological factors
* Brain chemistry and neural functioning
* Stress and motivation
* Physical health and exercise
* Music and lifestyle
* Social environment

---

### Cognitive Science Connection

Includes:

* Psychology
* Neuroscience
* Education
* Physiology
* Mathematics & computational modeling (neural networks)

## Conclusion

Effective learning depends on:

* Depth and quality of processing
* Use of higher cognitive skills (analysis, reflection, comprehension)
* Active engagement with learning materials

Brain-compatible strategies (20 listed in the lecture) improve memory, retention, and retrieval.
