# **Lecture 16: Thinking and Cognition**

## Introduction
- Today’s topic: **Thinking and Cognition**.
- Previously discussed: memory, learning, and different learning theories.
- **Thinking** is a cognitive activity, part of overall **cognition**.

---

## Cognition
- Cognition: mental processes involved in **gaining knowledge**, **processing information**, **remembering**, and **comprehending**.
- Includes **thinking, knowing, remembering, judging, problem-solving, summarizing, comprehending**, etc.
- Represents **higher-level brain functions**.
- Components of cognition:
  - Language
  - Imagination
  - Perception
  - Planning

---

## Thinking
- **Definition**: The process of consciously generating and manipulating ideas in the mind.
- **Importance**:
  - Solving problems
  - Learning new information
  - Understanding concepts
  - Organizing experiences
  - Communication, negotiation, judgment
- **Role**: Thinking is inseparable from learning and is the underlying mechanism for human cognition.

---

## Types of Thinking
1. **Creative Thinking**
   - Involves generating **new, original, unconventional ideas**.
   - Related to **intelligence** (basic grasping ability required).
   - Influenced by:
     - Imagination
     - Environmental inputs
     - Personality traits
   - Requires both **divergent and convergent thinking**.

2. **Divergent Thinking**
   - Generating **multiple solutions** to a problem.
   - Example: **Brainstorming**
   - Later stages involve **convergent thinking** to select the most effective solution.

3. **Symbolic Thinking**
   - Thinking in terms of **symbols, mental representations, or images**.
   - Example: Children playing with toys to replicate real-world roles.
   - Requires:
     - Active imagination
     - Stimulating environment
     - Curiosity
     - Intrinsic motivation and mastery in a field

---

## Concepts and Prototypes
- **Concepts**: Mental representation of a group or class of objects/events.
  - Reduce cognitive load
  - Organize information efficiently
- **Prototypes**: Representative example of a concept.
  - Example: Himalayan mountains as a prototype of "mountain".
- New information is compared with existing prototypes to understand or categorize it.

---

## Propositions
- Propositions: Meaningful statements formed through **manipulation of language, mental images, and ideas**.
- Help develop **mental models, frameworks, or prototypes** to organize knowledge.
- Essential for **strategic thinking**.

---

## Characteristics of Effective Thinking
- Active imagination
- Creative and stimulating environment
- Curiosity and inquisitiveness
- Expertise in a subject
- Intrinsic motivation

---

## Thinking as a Cognitive Process
- Uses both **working memory** and **long-term memory**.
- Involves:
  - Encoding information
  - Processing
  - Retaining
  - Transforming mental representations
- Symbolic in nature: uses signs, symbols, and language
- Can manifest through:
  - Art
  - Writing
  - Gestures
  - Sub-vocal/self-talk

---

## Thinking Across Lifespan
- Begins early in childhood:
  - Initial concepts often relate to caregivers.
  - Develops through **assimilation and accommodation**.
- Lifelong process:
  - Influenced by experience, exposure, and learning
  - Often **initiated by problems or objectives**

---

## Elements of Thought
1. **Mental Images**
   - Visual, kinesthetic, auditory, or olfactory
2. **Concepts**
   - Mental representation of classes or groups of things
3. **Prototypes**
   - Representative examples of a concept
4. **Propositions**
   - Structured manipulation of ideas to form meaningful relationships

---

## Strategic Thinking
- Input → Processing (strategies) → Output
- Enables flexible and effective problem-solving
- Uses all aspects of cognitive processing to achieve goals
