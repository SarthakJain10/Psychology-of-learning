### Motivation of Learning

## ‚ú® What is Motivation?

* **Cause of action** that drives a person to know, act, learn, and gain skills
* **Mental impulse** that energizes and directs learning behavior
* Makes a learner **keen, eager, mentally ready**
* Seen as **internal processes** revealed through:

  * Attention
  * Facial expressions
  * Engagement levels
  * Persistence in task

## üî• Types of Motivation

### 1) Intrinsic Motivation

* **Internal source of motivation**
* Arises from:

  * Personal interest
  * Curiosity
  * Satisfaction
  * Passion for mastering skills
* Leads to:

  * Deep learning
  * Better performance
  * Self-efficacy
  * Self-regulation

### 2) Extrinsic Motivation

* **External rewards or pressures**
* Examples:

  * Grades, certificates, prizes
  * Social pressure
  * Avoiding punishment
  * Financial incentives
* Includes shaping behavior through **operant conditioning**

## üîÑ Motivation Continuum (Self-Determination Theory)

Motivation is not binary. It flows:

1. **Amotivation**
   No intent or interest
2. **Extrinsic Motivation**
   Driven by rewards, avoidance of punishment
3. **Introjected Regulation**
   Internal pressure (guilt, social expectations)
4. **Identified Regulation**
   Learner recognizes value and importance
5. **Intrinsic Motivation**
   Pure inner desire and joy of learning

Peak: **Self-determination**
Combination of high intrinsic motivation and identified regulation

---

## üåü Why Intrinsic Motivation Matters More

* Promotes:

  * Deep conceptual understanding
  * Strong memory and performance
  * Confidence & independence
  * Creativity and innovation
* Key for **lifelong learning**

Extrinsic can **help start** the journey, but intrinsic **sustains** it.
