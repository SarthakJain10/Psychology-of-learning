# Lecture – 38: Learning and Pedagogy (Contd.)

## Critical Pedagogy

- Critical pedagogy is a part of social movements, encouraging learners to question power, social issues, inequalities, and hierarchies.
- Purpose: Develop critical awareness and responsible, active, global citizens.  

### Steps of Critical Pedagogy:
1. **Unlearning** – Forgetting rote memorization and traditional methods to prepare for rational thinking and social learning.
2. **Learning** – Acquiring new knowledge and skills.
3. **Relearning** – Revisiting important issues if required.
4. **Reflection** – Reflecting on learning experiences.
5. **Evaluation** – Critically assessing understanding and outcomes.

### Key Features:
- Student-centered classroom.
- Encourages critique of societal power structures.
- Teachers challenge inequalities in family, health, education, and society.
- Promotes understanding of cause-effect relationships, social injustices, and democratic values.

## Personalizing Pedagogy

- Utilizes information technology to provide **varieties of choices** in learning.
- Focus: Customizing learning as per learner’s needs, preferences, adaptability, and flexibility.
- Key aspects:
  - Learner-centered approach.
  - Self-directed learning.
  - Effective use of online platforms, discussion forums, and live sessions.
  - Teacher as facilitator, integrating student feedback for innovative pedagogy.

---

## Culturally Responsible Teaching

- Emphasizes respect for **cultural diversity** in classrooms.
- Ensures minority groups' perspectives and cultural identities are recognized and included.
- Goal: Equitable participation and comfort for all learners.

## Multimedia Learning

- Enhances learning by combining technology with research-based pedagogy.
- **Core assumptions (Mayer's Cognitive Theory of Multimedia Learning):**
  1. **Dual Channels:** Visual and auditory information are processed separately.
  2. **Limited Capacity:** Each channel has restricted processing ability.
  3. **Active Processing:** Learners actively construct coherent mental representations.

### Components:
- Words (auditory channel: spoken or printed)
- Pictures (visual channel: photos, illustrations, animations, graphs, videos)

### Steps in Multimedia Learning:
1. Select relevant words and images.
2. Organize words into verbal mental models.
3. Organize images into visual/pictorial mental models.
4. Integrate verbal and visual models with prior knowledge.
5. Facilitate meaningful learning for long-term retention.

### Multimedia Principles:
- **Multimedia Principle:** Words + pictures > words alone.
- **Spatial Contiguity Principle:** Corresponding words and pictures should be close together.
- **Temporal Contiguity Principle:** Present words and pictures simultaneously.
- **Coherence Principle:** Exclude extraneous words, pictures, and sounds.
- **Modality Principle:** Animation + narration > animation + on-screen text.
- **Redundancy Principle:** Avoid animation + narration + on-screen text.
- **Individual Differences Principle:** Design effects stronger for low knowledge learners.

### Research-Based Principles:
- **Personalization Principle:** Conversational style words > formal style.
- **Interactivity Principle:** Learners control pace of presentation.
- **Signaling Principle:** Words include cues to guide understanding.
- **Individual Differences Principle:** Adapt design effects to learner knowledge levels.

### Theoretical Foundations:
- **Sweller’s Cognitive Load Theory** – Managing intrinsic and extraneous cognitive load.
- **Paivio’s Dual Coding Theory** – Visual + auditory information coding.
- **Baddeley’s Working Memory Model** – How information is retained and processed.
