EQ is shaped by various *personality attributes* including:

* **Self-regard**
  Respecting and valuing oneself.
* **Emotional awareness**
  Understanding our own feelings and those of others.
* **Assertiveness**
  Ability to express beliefs confidently and say ‚Äúno‚Äù when needed.
* **Empathy**
  Stepping into others‚Äô shoes and understanding their experiences.
* **Interpersonal relationships**
  Building and maintaining healthy, supportive social connections.
* **Stress tolerance**
  The resilience and mental strength to handle pressure effectively.
* **Impulse control**
  Ability to think before acting and avoid hyper-reactivity.
* **Reality testing**
  Checking whether perceptions align with actual facts.
* **Flexibility**
  Being adaptable and open-minded instead of rigid or stubborn.
* **Problem-solving**
  Identifying causes and applying practical, stage-wise solutions.
* **Self-actualization**
  Achieving one‚Äôs potential and contributing to society and global wellbeing.
* **Independence**
  Being self-reliant in everyday functioning and responsibilities.
* **Optimism**
  Hopefulness grounded in realism, not false positivity.
* **Happiness**
  A calm mental state that supports creativity and growth.

> These components originally formed *EQ* but later became known as **facilitators** in Bar-On‚Äôs model.

### üìò Daniel Goleman‚Äôs Theory of Emotional Intelligence

Goleman (2001) expanded EQ into **five core domains**:

1. **Knowing your own emotions** (Self-awareness)
2. **Managing your emotions** (Self-regulation)
3. **Motivating oneself** (Intrinsic drive)
4. **Recognizing emotions in others** (Social awareness/empathy)
5. **Handling relationships** (Social skills)

#### Why EQ matters

EQ predicts:

* Academic achievement
* Psychological adjustment
* Workplace success
* Ability to cope with stress, culture, communication and pressure

> EQ is a **learnable ability** and can be developed at any stage of life.
