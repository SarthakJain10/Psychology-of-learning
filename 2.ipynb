## üß† Behaviorism ‚Äî Historical Perspective

* Shift from structuralism & functionalism ‚Üí **Behaviorism** (early 20th century)
* Founder: **J. B. Watson**

  * Strong believer in environmental influence on learning
  * Famous idea: Behavior can be shaped entirely by environment

### Key Ideas of Behaviorism

* Psychology should study:

  * Observable behavior
  * Measurable responses
* Behavioral change = evidence of learning
* Experimental methods used to make psychology ‚Äúscientific‚Äù

### Example: Little Albert Experiment

* Child conditioned to fear a white rat by pairing it with loud noise
* Demonstrated **classical conditioning** in human behavior
