## 1. Introduction
- Continuation from previous lecture on **critical thinking skills**.

---

## 2. Reflective Thinking Skills
- **Definition:** Reflective thinking involves self-assessment, self-regulation, and self-checking.  
- **Relation to other thinking types:**
  - Prerequisite for **critical thinking** and **creative thinking**.  
  - Closely related to **metacognitive thinking** (awareness and regulation of one’s own cognition).  
- **Competencies involved:**  
  - Organizing, reasoning, developing assumptions, predictive skills, problem-solving competencies.  
- **Development strategies:**  
  - Writing and learning texts, keeping reflective diaries or portfolios.  
  - Constructing **concept maps** (advance organizers).  
  - Self-checking questions, reflective discussions, negotiated learning, and debates.  

---

## 3. Creative Thinking
- **Definition:** Creative thinking is the process of being innovative, thinking out-of-the-box, and generating unique solutions.  
- **Related terms:**
  - **Creativity:** Broad concept including processes, products, mechanisms, and personal attributes.  
  - **Innovativeness:** Application of creative ideas in real-world contexts.  
  - **Creative Thinking:** The cognitive process of generating new ideas or solutions.  
- **Manifestations:**  
  - New approach to problems, conflict resolution, problem-solving, artistic expressions, and innovative solutions.  
- **Cognitive requirements:**  
  - Analytical thinking, problem-solving, organizational skills, communication, observation, and divergent/convergent thinking.  
- **Process of creative thinking:**  
  1. **Preparation:** Gather information from multiple sources.  
  2. **Incubation:** Process ideas actively and subconsciously.  
  3. **Insight:** Eureka moment of solution.  
  4. **Evaluation:** Experiment, test, and assess solutions.  
  5. **Elaboration:** Apply, narrate, and integrate into practice.  
- **Enhancement strategies:**  
  - Play puzzles and mental games, ask “how” and “why” questions, read extensively, travel, and interact with diverse people.  
  - Encourage risk-taking (calculated risks).  

---

## 4. Lateral Thinking
- **Definition:** Expanding thinking boundaries by collecting diverse information and allowing the mind to generate multiple options.  
- **Purpose:** To develop a **repository of ideas** that can be used when required.  

---

## 5. Problem Solving
- **Definition:** Solving assigned or encountered problems using systematic steps.  
- **Strategies:**  
  - **Trial and error:** Gradual error correction.  
  - **Algorithm:** Applying known formulas or procedures.  
  - **Analogy:** Applying solutions from similar past problems.  
- **Relation to creative thinking:**  
  - Problem-solving is often enhanced by creative thinking for generating multiple potential solutions.  

---

## 6. Problem-Based Learning (PBL)
- **Definition:** Experiential and constructive learning approach where learners solve real-life problems.  
- **Characteristics:**  
  - Learners can **select or be assigned problems**.  
  - Emphasizes experiential learning, real-life application, and skill development.  
  - Integrates **cognitive, affective, and psychomotor skills**.  
- **Learning cycle in PBL:**  
  1. Determine what is known and what needs to be known.  
  2. Sequence thoughts and gather information.  
  3. Generate potential solutions and formulate hypotheses.  
  4. Apply new knowledge to the problem.  
  5. Evaluate and discuss outcomes.  
  6. Peer collaboration, sharing perspectives, and critical evaluation.  
- **Role of learner:** Central, responsible for problem identification, strategy, evaluation, and presentation.  

---

## 7. Integration of Thinking Skills
- **Reflective, critical, metacognitive, and creative thinking** are interconnected.  
- PBL and collaborative learning facilitate **cross-fertilization of ideas** and innovation.  
- Continuous practice and nurturance enhance **confidence, scientific thinking, perseverance, and lifelong learning skills**.  

---

## 8. Key Takeaways
- Reflective thinking → prerequisite for higher-order thinking.  
- Creative thinking → systematic cognitive process, enhances innovation.  
- Problem-solving → combines creativity and systematic strategies.  
- PBL → practical application of all thinking skills in experiential learning contexts.  
- Nurturing thinking skills adds **value to learning, performance, and problem-solving capacity**.
