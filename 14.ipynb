## 1. Memory Types in Long-Term Memory (LTM)

### 1.1 Declarative Knowledge
- **Definition:** Knowledge of facts, numbers, dates, names, and verbal data.
- **Characteristics:**
  - "Know what"
  - Extensive capacity
  - Explicit memory, can be consciously recalled

### 1.2 Procedural Knowledge
- **Definition:** Knowledge of processes, methods, and skills.
- **Characteristics:**
  - "Know how"
  - Includes problem-solving strategies, formulas, lab experiments, mnemonic methods
  - Implicit memory, usually not consciously recalled
  - Learned over years of education and practice

### 1.3 Episodic Knowledge
- **Definition:** Memory of specific life events or episodes.
- **Examples:**
  - Graduation day
  - Personal experiences with emotional relevance
- **Characteristics:**
  - Autobiographical episodic memory: specific to one’s life events
  - Experimental episodic memory: learning facts or experiments
  - Flash memory: key vital points of an event

### 1.4 Semantic Memory
- **Definition:** Memory of meanings and concepts derived from declarative knowledge.
- **Examples:** Gender equality, democracy, feminism
- **Characteristics:**
  - Meaning-based encoding
  - Derived from episodic or declarative knowledge
  - Explicit memory

## 3. Level of Processing Theory (Craik & Lockhart)
- **Idea:** Depth of information processing affects retention in LTM.
- **Types of Processing:**
  1. **Shallow Processing**
     - Structural encoding (physical appearance)
     - Example: Rote memorization before exams
     - Results in poor retention
  2. **Intermediate Processing**
     - Phonemic/phonetic encoding (sound-based)
     - Example: Rhyming, auditory cues
  3. **Deep Processing**
     - Semantic encoding (meaning-based)
     - Elaborative rehearsal, reflection, analysis
     - Example: Concept maps, associations, critical evaluation
     - Results in better recall and long-term retention

- **Key Insight:** Deeper processing → more accurate and durable memory.

- Cognitive neuroscience approach
- Memory capacity influenced by:
  - Synapse complexity
  - Sparseness of neural representation
  - Spatial and temporal correlations
- Neural networks integrate signals, forming memory structures
- Highlights biological basis of memory but not explored in depth in this lecture

## 6. Cognitive Load
- **Definition:** Mental workload during information processing.
- **Cognitive Architecture:**
  - Memory system (sensory, working, LTM)
  - Schemas (pre-existing concepts and structures)
- **Schema Theory:**
  - Knowledge stored in LTM in organized schemata
  - Allows integration of large amounts of information as a single unit
  - Automated schemata reduce storage and processing load
  - Skilled performance achieved through complex, flexible schema integration

---

## 7. Key Points
- Declarative, procedural, and episodic memories form the backbone of LTM.
- Level of processing influences how well information is retained.
- Cognitive load can be managed through schema-based organization and automation.
- Deep, elaborative processing enhances memory retention and recall.
- Human memory processing can be compared with computer models for better understanding.
