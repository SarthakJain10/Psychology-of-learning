# Major Theories of Learning

## 1. Introduction

* Focus on **major learning theories** and **schools of thought**
* Three primary approaches:

  1. **Behavioristic School**
  2. **Cognitive Approach**
  3. **Constructivist Approach**
* Behaviorism historically dominated by American psychologists

---

## 2. Behaviorism: Core Ideas

* Founded by **John B. Watson**, later developed by **Pavlov**, **Skinner**
* Psychology viewed as:

  * A **science** of observable and measurable behavior
  * Mentalistic concepts (mind/soul) excluded
* Learning defined as:

  * Forming **Stimulus–Response (S–R) connections**
  * Behavior shaped through **environmental stimuli**
* Major contributions:

  * Introduction of **experiments**, use of **animals**
  * Focus on measurable outcomes

---

## 3. Limitations of Behaviorism

* Cannot explain:

  * Internal cognitive processes
  * Some human behaviors (complexity of children’s responses)
* Only observable behavior considered valid

---

## 4. Classical Conditioning – Ivan Pavlov

### Key Concepts

| Term                            | Meaning                                                      |
| ------------------------------- | ------------------------------------------------------------ |
| **US (Unconditioned Stimulus)** | Naturally produces a response (e.g., food)                   |
| **UR (Unconditioned Response)** | Natural response (e.g., salivation)                          |
| **CS (Conditioned Stimulus)**   | Previously neutral stimulus becoming associated (e.g., bell) |
| **CR (Conditioned Response)**   | Learned response to CS (e.g., salivation at bell)            |

### Process

* Bell (neutral) paired with food (US)
* Association formed through repetition
* Bell alone ⇒ salivation (CR)
* **Learning through association**

### Timing Variations in Conditioning

| Type                          | Description                                                  |
| ----------------------------- | ------------------------------------------------------------ |
| **Delayed Conditioning**      | CS presented, US follows with small overlap (most effective) |
| **Trace Conditioning**        | CS ends before US starts; gap present                        |
| **Simultaneous Conditioning** | CS and US occur together                                     |
| **Backward Conditioning**     | US before CS (least effective)                               |

---

## 5. Principles of Classical Conditioning

* **Acquisition**
  Formation of CS–CR association
* **Extinction**
  CS repeated without US ⇒ response weakens
* **Spontaneous Recovery**
  After pause, extinguished response reappears
* **Stimulus Generalization**
  Similar stimuli evoke same response
  (e.g., different bell tones)
* **Stimulus Discrimination**
  Organism learns differences between stimuli
* **Higher-Order Conditioning**
  CS paired with new neutral stimulus forming new CS
  (e.g., bell + light + food)

---

## 6. Operant Conditioning – B. F. Skinner

* Learning through **consequences**
* Behavior shaped by rewards and punishments

### Experiment Example

* Rat presses lever → receives food
* Learns which actions bring **positive outcomes**

### Key Terms

| Concept           | Meaning              |
| ----------------- | -------------------- |
| **Reinforcement** | Strengthens behavior |
| **Punishment**    | Weakens behavior     |

### Types of Reinforcement

| Type                       | Effect                     | Example                          |
| -------------------------- | -------------------------- | -------------------------------- |
| **Positive Reinforcement** | Add pleasant stimulus      | Rewards for good grades          |
| **Negative Reinforcement** | Remove unpleasant stimulus | Avoiding shock by correct action |

### Types of Punishment

| Type                    | Effect                 | Example                  |
| ----------------------- | ---------------------- | ------------------------ |
| **Positive Punishment** | Add unpleasant stimuli | Scolding for misbehavior |
| **Negative Punishment** | Remove desirable item  | Taking away privileges   |

---

## 7. Key Takeaways

* Behavior is **conditioned** through environment and consequences
* Behavior can be **learned**, **modified**, **unlearned**
* Widely applied in:

  * Child behavior shaping
  * Classroom management
  * Training and behavior therapy
