## **Maslow’s Hierarchy of Needs**

Maslow proposed that human needs are structured in a **hierarchical pyramid**, where lower-level needs must be satisfied before higher-order desires emerge.

### **5-Tier Model**

1. **Physiological Needs**

   * Food, clothing, shelter, sleep
   * Basic biological survival requirements

2. **Safety Needs**

   * Job security, financial stability
   * Protection from threats and uncertainty

3. **Love and Belongingness Needs**

   * Affiliation, friendships, family bonds
   * Social acceptance and emotional connection

4. **Esteem Needs**

   * Status, recognition, respect, achievement
   * Sense of competence and self-worth

5. **Self-Actualization**

   * Realizing full potential, creativity, purpose
   * Working toward values beyond the self
   * Examples: innovators, reformers, humanitarians

### **Later Addition**

* **Transcendence / Transcendental Needs**

  * Spiritual growth, thinking beyond life and self
  * Service to humanity and future generations

---

### **Types of Needs**

| Type                 | Levels                                       | Focus                                         |
| -------------------- | -------------------------------------------- | --------------------------------------------- |
| **Deficiency Needs** | Physiological, Safety, Belongingness, Esteem | Driven by deprivation                         |
| **Growth Needs**     | Self-Actualization (and Transcendence)       | Driven by desire to grow, create, help others |

### **Key Insights**

* Motivation progresses from survival to growth.
* Theory widely applied in education, organizations, workplace well-being.
* Explains how motivation evolves with fulfilled basic needs.

## **Self-Determination Theory (SDT)**

*Proposed by Ryan & Deci (2000)*

This theory connects:

* **Human motivation**
* **Personality**
* **Optimal functioning**

### **Core Psychological Needs**

| Need            | Description                              |
| --------------- | ---------------------------------------- |
| **Autonomy**    | Feeling in control of one’s own actions  |
| **Competence**  | Feeling effective and capable            |
| **Relatedness** | Feeling connected and supported socially |

When these needs are fulfilled, motivation, engagement, and performance increase.

---

### **Types of Motivation in SDT**

Motivation exists on a **continuum**, from non-self-determined to fully self-determined:

| Motivation Type                         | Regulation                            | Drivers                                      |
| --------------------------------------- | ------------------------------------- | -------------------------------------------- |
| **Amotivation**                         | No regulation                         | Lack of intention or interest                |
| **External Regulation** (Controlled)    | External pressure                     | Rewards or punishments                       |
| **Introjected Regulation** (Controlled) | Partially internal                    | Avoid guilt, seek approval or ego protection |
| **Identified Regulation** (Autonomous)  | Conscious valuing                     | Personally important goals                   |
| **Integrated Regulation** (Autonomous)  | Internalized values aligned with self | Behavior fits identity and beliefs           |
| **Intrinsic Motivation**                | Inherent enjoyment                    | Interest, curiosity, personal satisfaction   |

---

### **Autonomous vs Controlled Motivation**

| Autonomous Motivation                      | Controlled Motivation                 |
| ------------------------------------------ | ------------------------------------- |
| Comes from inner desire and aligned values | Driven by fear, duty, pressure        |
| Increases self-direction and satisfaction  | Decreases autonomy and internal drive |

---

### **Major Takeaways**

* Motivation is **not binary**; it **develops gradually** like a dimmer switch rather than an ON/OFF button.
* Self-determined motivation leads to better learning, creativity, and well-being.
* Human motivation grows as internal values and identity become stronger.
