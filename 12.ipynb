### Atkinson & Shiffrin Model

Memory involves three major stages:

1. **Encoding**
2. **Storage**
3. **Retrieval**

Information flow:

```
Environment ‚Üí Sensory Memory ‚Üí Short-Term/Working Memory ‚Üí Long-Term Memory
                                       ‚Üë                       ‚Üì
                                   Retrieval ‚Üî Rehearsal
```

Memory processing is **bidirectional**. Working memory retrieves stored information from LTM while solving tasks.

## ‚ùå Forgetting

Occurs due to:

* Encoding failure
* Retrieval failure
* Interference of new/old info
* **Decay** (lack of use)
* Brain damage

Most forgetting happens within **4 weeks** if no rehearsal follows learning.

Greater forgetting:

* When subject matter is **unfamiliar**
* When learning is **surface-level**

---

## üéì Educational Implications

To improve memory and learning:

* Encourage **active rehearsal**
* Use meaningful organization and chunking
* Promote engagement through:

  * Group work
  * Discussion & debate
  * Concept mapping
  * Real-life connections
* Focus on higher order thinking:

  * Problem-solving
  * Metacognition
  * Reflection

Better engagement = Better retention ‚úÖ
