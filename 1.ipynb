## üéØ What Is Learning?

### Common Misconception:

Learning is only about performance, marks, grades, or exam results.

### Psychological Perspective:

Learning:

* **Brings a relatively permanent change in behavior**
* Results from **experience, practice, or training**
* Change is **observable** and meaningful

### Key Understanding:

Learning = **Change + Experience + Persistence**

Example:

> When training a pet, visible behavioral improvements indicate learning
> although not always permanent or perfect.

---

## üîç Conditions to Confirm Learning

Two requirements must be met:

1. ‚úÖ **Observable change in behavior**
2. ‚úÖ Change is **caused by experiences or environment**

Learning is seen as an **effect** (outcome) of interactions with environment or stimuli.

## üß© Cognitive View of Learning

* Learning changes **mental constructs**: ideas, schemas, associations
* Changes in behavior occur **because** thought patterns evolve
* Even physical brain structure can change
  (neural networks & brain chemistry reshape through learning)

---

## üåç Principles That Influence Learning

### 1. Environment Matters

* School, home, work, community surroundings influence learning ability
* A rich, supportive environment boosts performance
  (resources, emotional support, interactions)

Ongoing debate:
**Nature (heredity/genetics) vs. Nurture (environment)**
Both contribute to learning, degree varies.

### 2. Reinforcement

* Positive feedback, praise, rewards, collaboration
* Stimulates motivation and continued learning

Examples of reinforcement:

* Teacher feedback
* Scholarships or awards
* Peer recognition

### 3. Learning through Association & Connection

Learners:

* Link new information with **past experiences**
* Build networks of meaning between concepts

To learn effectively:

* Facts + Concepts + Interpretations must be stored and **connected** in memory
* Retrieval of old knowledge helps understand new ideas

---

## ‚úÖ Summary

Learning is:

* An **enduring** change
* Caused by **experience, environment, reinforcement, and cognition**
* Supported by changes in **brain structure and mental pathways**
* Deeply influenced by **how learners connect prior knowledge** to new content
