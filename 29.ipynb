# *Lecture 29: Emotional Intelligence and Learning (Contd.)*


### ✅ **Emotional Labour**

#### Meaning

* Effort involved in managing and expressing emotions **as required by the situation or organization**.
* Moving emotional management from **private** (home/personal life) to **public** (workplace/classroom).

> “Displaying emotions we may not actually feel, because the role demands it.”

#### Features

* Managing **inner feelings** to show socially acceptable emotions.
* Goal: Ensure others feel comfortable (e.g., clients, coworkers).
* Culture and situation specific.
* Often required in **group settings**, professional environments.

#### Where is emotional labour high?

* Hospitality industry (hotels, restaurants)
* **Flight attendants** (classic example)
* Customer service roles
* Healthcare and education
* Any role with **direct public interaction**

#### Organizational rules

* Some workplaces restrict display of personal emotions (e.g., crying considered *unprofessional*).
* Feelings should not disrupt work culture or influence others.

#### Gender aspect

* Jobs with high emotional labour often **gender-typed as female roles**

  * Receptionists, cabin crew, support roles
* Society assumes women are better at emotional warmth and social harmony.

---

### ⚙️ Types of Emotional Labour

| Type               | Description                                                  | Example                                        | Effect                                                           |
| ------------------ | ------------------------------------------------------------ | ---------------------------------------------- | ---------------------------------------------------------------- |
| **Surface Acting** | Changing outward expressions without feeling them internally | Smiling despite illness or stress              | Leads to exhaustion, burnout, depersonalization, dissatisfaction |
| **Deep Acting**    | Trying to actually feel the expected emotions                | Genuinely empathizing with someone in distress | Positive emotional performance, personal fulfilment              |

---

### 🎓 Emotional Labour in Teaching and Learning

* Teaching requires **both cognitive and emotional effort**.
* Teachers must maintain a **positive and composed demeanor** while handling:

  * Student expectations
  * Management pressure
  * Parent demands
* High accountability. Many stakeholders.
* Continuous emotional regulation can lead to:

  * Burnout
  * Exhaustion
  * Decline in mental health

#### Effects:

* **Surface acting** in teaching → more burnout & dissatisfaction
* **Deep acting** → enhances empathy, student support & personal accomplishment

#### Importance of strategies

* Healthy emotional regulation strategies protect:

  * Wellbeing
  * Job satisfaction
  * Classroom environment

---

### 🌐 Emotional Contagion

#### Definition

* When one person’s emotions trigger similar emotions in others.
* Spread occurs through **observation, mimicry, group influence**.

#### Examples

* Laughing spreads quickly in a group
* Smiling at someone → they tend to smile back
* Violence in crowds or riots escalates due to emotional contagion
* Children imitate emotional expressions from peers or adults

> Both **positive and negative emotions** are contagious.

---

### 📌 Key Takeaways

* Emotional labour extends personal emotion regulation into **public roles**.
* Required more in roles involving **regular social interaction**.
* Must be balanced with organizational support to prevent **burnout**.
* Emotional contagion highlights how group emotions influence each other.
