# Lecture 27 – Emotional Intelligence and Learning (Contd.)

## Emotional Intelligence (EI): A Learnable Ability

* EI is **not fixed**. It can be developed at **any age**.
* Though temperament is innate, EI grows with:

  * Learning
  * Social experiences
  * Self-regulation practice

---

## Advantages of Developing Emotional Intelligence

* Accepting **self-criticism** and responsibility
* Ability to **move on** after mistakes and correct them
* Knowing when and how to **say “No”**
* **Sharing feelings** appropriately to build trust and bonding
* **Problem-solving** that works for everyone (consensus decisions)
* Showing **empathy** for others’ problems and setbacks
* Practicing **active listening**:

  * Builds trust, reduces conflicts
  * Shows care and concern
* Being aware of **reasons behind one's actions**
* Being **non-judgmental** and avoiding premature biases

---

## EI as a Competency

EI supports:

1. **Personal Competence**

   * **Self-awareness**

     * Knowing emotional states, strengths, weaknesses
     * Realizing how behavior affects others
     * Avoiding emotional manipulation by others
   * **Self-management**

     * Regulating impulses and disruptive emotions
     * Acting in line with personal and organizational values
     * Flexibility and persistence despite obstacles

2. **Social Competence**

   * **Social Awareness**

     * Reading group moods and body language
     * Understanding emotional impacts in social settings
   * **Relationship Management**

     * Building rapport, especially with new members
     * Handling conflicts effectively
     * Communicating clearly
     * Empathy and sensitivity
     * Motivating and influencing others

---

## EI in Careers & Social Contexts

* Plays a key role in **marketing and customer-facing roles**:

  * Hospitality, hotel management
  * Receptionists, flight attendants
  * Health and service sectors
* Marketers strategically use customer emotions to persuade purchases

---

## Steps in Emotional Processing

(Sequential developmental nature)

1. **Perceiving Emotions**

   * Understanding non-verbal cues and expression
2. **Reasoning with Emotions**

   * Emotions guide thinking and decision priorities
3. **Understanding Emotions**

   * Interpreting complex emotional meanings
   * Emotional vocabulary + reasoning
4. **Managing Emotions**

   * Handling stress, crisis, and others’ emotions effectively
   * Staying rational and solution-oriented

---

## Emotional Intelligence & Personality

* Personality = **psychophysical traits** shaped by:

  * Socialization, learning, environment, values
* EI and cognitive intelligence both contribute significantly to personality

---

## Gardner’s Emotional Intelligence Pyramid (2011)

Nine-layer hierarchical model combining:

* Interpersonal + intrapersonal intelligence
* Ability + trait EI models

Layers include:

1. Emotional stimuli
2. Emotional regulation and perception
3. Expression of emotions
4. Self-awareness
5. Self-management
6. Social awareness
7. Empathy and discrimination of emotions
8. Emotional expertise & universality (love, care, empathy)
9. Self-actualization → Emotional unity (transcendence)

---

* EI helps resolve:

  * Task/process conflicts
  * Especially **relationship-based conflicts**
