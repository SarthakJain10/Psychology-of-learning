## ‚úÖ **Expectancy Theory of Motivation (Recap & Details)**

Motivation is influenced by the **expectation** that effort ‚Üí performance ‚Üí reward.
We make internal judgments before deciding to pursue a goal.

### Factors considered:

* **Past experience**
* **Self-confidence in ability**
* **Perceived difficulty of the task**
* **Control over overcoming challenges**

### Core Components:

| Component           | Meaning                                  | Key Questions                              |
| ------------------- | ---------------------------------------- | ------------------------------------------ |
| **Expectancy**      | Belief that effort leads to performance  | ‚ÄúCan I do this if I work hard?‚Äù            |
| **Instrumentality** | Belief that performance will be rewarded | ‚ÄúWill my performance be fairly evaluated?‚Äù |
| **Valence**         | Perceived value of the reward            | ‚ÄúDo I really want this reward?‚Äù            |

### Instrumentality Influences:

* Reward criteria transparency
* Performance appraisal system fairness
* Preparedness for the task
* Difficulty level vs skill match

### Valence Outcomes:

* Can be **positive**, **neutral**, or **negative**
* Depends on personal priorities and trade-offs
  (e.g., promotion vs work-life balance)

## üéØ **Goal Setting Theory (Edwin Locke)**

Motivation increases when **clear and meaningful goals** are set.

### Effective Goal Characteristics:

1. **Clarity**
   Goals must be specific, not vague or overly broad.
2. **Challenge**
   Slight difficulty increases effort and engagement.
3. **Attainability**
   Ambitious but achievable goals prevent frustration.
4. **Acceptance**
   Self-chosen goals yield stronger motivation than imposed goals.
5. **Feedback**
   Continuous feedback ensures correct progress.
6. **Performance-Focused**
   Goals should improve performance, not just image.
7. **Deadlines**
   Increase accountability and timely effort.

## üß† **Learning Goals vs Performance Goals**

| Learning Goal Orientation                | Performance Goal Orientation      |
| ---------------------------------------- | --------------------------------- |
| Focus on learning, mastery, skill growth | Focus on results, rewards, grades |
| Higher intrinsic motivation              | More extrinsic motivation         |
| Promotes innovation & resilience         | May cause anxiety during failures |
