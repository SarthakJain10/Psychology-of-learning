# Lecture – 34: Learning and Instruction (Contd.)

## ADDIE Model (Instructional System Design)

- **Purpose:** Framework to design effective learning, training, and development programs.
- **Phases:**  
  1. **Analysis** – Set goals, research target audience, identify requirements.  
  2. **Design** – Develop learning solutions, objectives, content focus.  
  3. **Development** – Create course content, integrate technology, prepare resources.  
  4. **Implementation** – Execute training, deliver content, conduct sessions.  
  5. **Evaluation** – Assess effectiveness, formative and summative evaluation.

### Analysis Phase
- Key Questions:
  - Who is the audience? (IT professionals, doctors, engineers, trainees)
  - What are their characteristics?
  - What are the expected new behavioral outcomes?
  - Learning constraints (infrastructure, financial, technology)
  - Delivery options (online/offline, pedagogical considerations)
  - Timeline of the program
- Types of Analysis:
  - **Need Analysis:** Existing vs. required knowledge and skills
  - **Financial Analysis**
  - **Situational & Contextual Analysis:** Resources, infrastructure, constraints
  - **Pedagogical Analysis:** Suitable teaching methods
  - **Time Analysis** and other logistical considerations

### Design Phase
- Activities:
  - Formulate learning/training/instructional objectives
  - Select assessment instruments, exercises, content sources
  - Lesson planning and media selection
  - Documentation of design blueprint (instructional, technical, visual strategies)
  - Storyboards, user interface/experience, prototype creation
  - Pedagogical approaches (case studies, hands-on experience, software integration)

### Development Phase
- Activities:
  - Assemble content and resources
  - Integrate required technology
  - Revise according to feedback
  - Organize schedule, expert sessions, hospitality, logistics
  - Prepare materials, software, labs, websites, hands-on equipment

### Implementation Phase
- Actual delivery of the program
- Procedures include:
  - Training facilitators
  - Course execution
  - Feedback mechanism for iterative improvements
  - Hands-on exercises, field visits, online/offline delivery
  - Alignment of assessments with instructional objectives

### Evaluation Phase
- **Formative Evaluation:** Continuous assessment during the program
  - Quizzes, participation, module-wise exercises
- **Summative Evaluation:** Final assessment at the end of the program
  - Tests, certification, grading
- Purpose: Ensure learning objectives are met and knowledge is transferable

## Instructional Framework

- **Purpose:** Illustrates interrelationships among:
  - Instructional approaches
  - Instructional models
  - Instructional strategies
  - Instructional methods
  - Instructional skills
- **Goal:** Align strategies, methods, and skills with learning objectives
- **Customization:** Adjusted per audience, course, and training goals

### Components
1. **Instructional Models:** Holistic framework, e.g., Joyce & Weil:
   - **Information Processing Model**
   - **Behavioral Model**
   - **Social Interaction Model**
   - **Personal Model**
2. **Instructional Strategies:** Direct, indirect, interactive, experiential, independent study
3. **Instructional Methods:** Lecture, case study, cooperative learning, simulation, debate, discussion, flipped/blended learning
4. **Instructional Skills:** Planning, presenting, demonstrating, questioning, evaluating

---

## Instructional Models (Joyce & Weil)

### Information Processing Model
- Focus on **cognitive development** and individual self-concept
- Emphasizes understanding, organizing, and processing information
- Encourages personal and societal relationships

### Behavioral Model
- Based on **behaviorist learning theory**
- Focus on observable behavior changes
- Instructional tasks sequenced in small steps
- Decision-making linked to student assessment, prior knowledge, and learning styles

## David Ausubel’s Instructional Theory

- **Focus:** Meaningful learning
- Learning is **integrated into existing knowledge structures** (mental schemas)
- **Prior knowledge is critical**
- **Advanced Organizer:** Primer/orientation for new learning
  - Facilitates understanding of complex material
  - Prepares learners mentally for new content
- **Types of Organizers:**
  1. **Comparative Organizer:** Activates prior knowledge, recapitulates, integrates or discriminates new info
  2. **Expository Organizer:** Introduces new, unfamiliar material
- **Key Principles:**
  - Knowledge is **hierarchically organized**
  - Learning occurs through **subsumption**
    - New material related to existing cognitive structure
    - Includes assimilation, accommodation, and cognitive restructuring
- **Outcome:** Hierarchical organization of knowledge from simple to complex

---

### Summary

- ADDIE model provides **stepwise instructional design**
- Instructional framework links **models, strategies, methods, and skills**
- Ausubel emphasizes **meaningful learning** using **advanced organizers**
- Subsumption facilitates integration of **new knowledge with existing schemas**
