## Cognitive Architecture
- Serves as a **computational model of the brain**, often used in cognitive science and computer science.  
- Helps understand:  
  - Brain structure and components  
  - How the mind functions  
  - How cognitive processes operate  

### Key Points:
1. **Brain as a Computer Analogy**  
   - Memories: working memory, long-term memory, semantic and episodic memory  
   - Activation mechanisms  
   - Learning mechanisms: symbolic, statistical, algorithmic  
   - Human limitations and perceptual-motor parameters  
   - Interaction with environment: input/output  

2. **Purpose**  
   - Understand how the **brain architecture activates the mind** to perform tasks.  
   - Illustrates that **human brain functions in structured, computational ways**.

## Cognitive Hierarchy
- **Cognition develops hierarchically** from simple to complex through life experience and learning.  
- Different models exist for organizing cognitive processes:  
  - **Information processing speed**  
  - **Attention and concentration**  
  - **Perception**  
  - **Working memory and long-term memory**  
  - **Executive functions**

## Bloom's Taxonomy of Learning Objectives
- Developed by **Benjamin Bloom (1956)**, revised later (2001).  
- **Three domains of learning**:  
  1. **Cognitive** – Intellectual abilities and knowledge  
  2. **Affective** – Emotion, motivation, values  
  3. **Psychomotor** – Physical movement, skills, performance  

### Cognitive Domain (Revised Taxonomy)
1. **Remember** – Recall facts, basic concepts  
2. **Understand** – Explain ideas, interpret concepts  
3. **Apply** – Use knowledge in new situations  
4. **Analyze** – Examine, compare, distinguish components  
5. **Evaluate** – Assess strengths, weaknesses, utility  
6. **Create** – Generate new ideas, innovate, construct original work  

- **Action verbs** (describe, relate, tell) help measure learning outcomes at each level.  

---

### Affective Domain (Bloom, 1964)
- Focus: Emotions, values, appreciation, motivation  
- Hierarchy:  
  1. **Receiving** – Awareness of stimuli  
  2. **Responding** – Reaction or participation  
  3. **Valuing** – Importance attached to information  
  4. **Organizing** – Prioritize and structure values  
  5. **Characterizing** – Internalize values into behavior  

---

### Psychomotor Domain
- Focus: Physical skills, competencies, mastery  
- Hierarchy of skill development:  
  1. **Imitation** – Observe and copy  
  2. **Manipulation** – Practice and follow instructions  
  3. **Precision** – Achieve expert performance  
  4. **Articulation** – Integrate multiple skills effectively  
  5. **Naturalization** – Skills become automatic and effortless  

---

## Types of Knowledge
1. **Factual Knowledge** – Basic data and information  
2. **Conceptual Knowledge** – Principles, categories, theories  
3. **Procedural Knowledge** – Skills, techniques, methods  
4. **Metacognitive Knowledge** – Awareness of one's own learning and thinking processes  

- **Metacognition**: "Learning about our own learning" – crucial for self-regulation and cognitive development.  

---

## Summary
- Cognitive architecture provides a **framework for understanding brain function** and the mind.  
- Bloom’s Taxonomy links **cognition, emotion, and skill acquisition** to structured learning objectives.  
- Metacognitive knowledge is the **highest-order cognitive ability**, guiding self-awareness and lifelong learning.
