Memory is a **psychological process** that enables us to:

1. **Acquire** information
2. **Store** information
3. **Retain** information
4. **Retrieve** information when needed

Example: Preparing for and writing exams.

## üéØ Goals of Human Memory

Memory involves:

* Preserving information for future use
* Recovering stored information when required

## üß¨ Memory and the Brain

Cognitive neuroscientists explain memory through **neural network changes**:

* Learning creates **new neural connections**
* Related neurons form **patterns or networks**
* Practice strengthens these synaptic connections
* Without rehearsal, memories fade

Learning and memory both involve **restructuring brain circuits**.

## üß© Role of Rehearsal & Mnemonics

* Reviewing, practicing, summarizing improves encoding
* Helps retain information long-term
* Strengthens synaptic links
* Techniques include:

  * Chunking
  * Visual cues
  * Color coding
  * Repeated revision

Mnemonic strategies are memory boosters.
